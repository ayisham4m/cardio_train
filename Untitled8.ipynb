```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
from matplotlib import pyplot as plt
import os
df = pd.read_csv("/content/drive/MyDrive/MP1B/cardio_train.csv", sep=";")
```

```python
fig, ax = plt.subplots()
for gender, group in df.groupby('gender'):
    ax.scatter(group['height'], group['weight'], label=gender)

plt.legend(loc='upper left')
plt.title('Tinggi Badan vs Berat Badan Berdasarkan Jenis Kelamin')
plt.xlabel('Tinggi Badan (cm)')
plt.ylabel('Berat Badan (kg)')
plt.grid(True)
plt.show()
```

```python
X_train = np.array(df[['height', 'weight']])
y_train = np.array(df['gender'])
print('X_train:\n', X_train)
print('y_train:\n', y_train)
```

```python
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
print('y_train:\n', y_train)
```

```python
y_train = y_train.flatten()
print('y_train:\n', y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier

K = 3
model = KNeighborsClassifier(n_neighbors=K)
model.fit(X_train, y_train)
```

```python
def create_data_point(height, weight):
  return np.array([[height, weight]])

X_new = create_data_point(170, 60)
print('X_new:\n', X_new)
```

```python
y_new = model.predict(X_new)
print('y_new:\n', y_new)
```

```python
from scipy.spatial import distance
import numpy as np
import pandas as pd

# Misalnya `misterius` dan `df` didefinisikan
misterius = np.array([1, 2])  # Ubah sesuai data Anda
df = pd.DataFrame({
    'heigth': [170, 176, 184],
    'weigth': [54, 51, 68],
})  # Contoh DataFrame

# Fungsi Euclidean
def euc(misterius, b):
    return distance.euclidean(misterius, b)

# Hitung jarak
df['distance'] = df.apply(lambda row: euc(misterius, row.values), axis=1)

# Urutkan berdasarkan jarak
df.sort_values('distance', inplace=True)
print(df)
```

```python
X_test = np.array([[168, 65], [180, 90], [166, 52], [169, 67]])
y_test = lb.transform(np.array(['male', 'male', 'female', 'female'])).flatten()

print(f'X_test:\n{X_test}')
print(f'y_test:\n{y_test}')
```

```python
Y_predicted = model.predict(X_test)
print(f'Y_predicted:\n{Y_predicted}')
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, Y_predicted)

print(f'Accuracy: {acc}')
```

```python
from sklearn.metrics import precision_score

precision = precision_score(y_test, Y_predicted)

print(f'Precision: {precision}')
```

```python
from sklearn.metrics import recall_score

recall = recall_score(y_test, Y_predicted)

print(f'Recall: {recall}')
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, Y_predicted)

print(f'F1 Score: {f1}')
```

```python
from sklearn.metrics import classification_report

cls_report = classification_report(y_test, Y_predicted)

print(classification_report(y_test, Y_predicted))
```

```python
from sklearn.metrics import matthews_corrcoef

mcc = matthews_corrcoef(y_test, Y_predicted)

print(f'MCC: {mcc}')
```
